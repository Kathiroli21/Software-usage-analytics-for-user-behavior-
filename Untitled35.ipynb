```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical

```

```python
data = pd.read_csv(r"D:\soil-paper\user_behavior_dataset.csv")
data
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='User Behavior Class', data=data, palette='viridis')
plt.title('Distribution of Target Classes', fontsize=16)
plt.xlabel('User Behavior Class')
plt.ylabel('Count')
plt.show()
```

```python
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
data = pd.get_dummies(data, columns=['Operating System', 'Device Model'], drop_first=True)
```

```python
data
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap of Features', fontsize=16)
plt.show()
```

```python
X = data.drop(columns=['User ID', 'User Behavior Class'])
y = data['User Behavior Class']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
y_encoded = to_categorical(y - 1, num_classes=5)  # Assuming classes are 1 to 5
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    layers.Dense(128, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=l2(0.01)),
    layers.Dropout(0.2),
    layers.Dense(64, activation='relu', kernel_regularizer=l2(0.01)),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu', kernel_regularizer=l2(0.01)),
    layers.Dropout(0.2),
    layers.Dense(5, activation='softmax')
])
```

```python
optimizer = Adam(learning_rate=0.005)

model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), callbacks=[early_stopping])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Training Accuracy', color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='orange')
plt.title('Training vs Validation Accuracy', fontsize=16)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.title('Training vs Validation Loss', fontsize=16)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

```

```python

test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

```

```python
feature_weights = np.mean(np.abs(model.layers[0].get_weights()[0]), axis=1)
sorted_indices = np.argsort(feature_weights)[::-1]

plt.figure(figsize=(10, 6))
plt.barh(np.array(X.columns)[sorted_indices], feature_weights[sorted_indices], color='teal')
plt.title('Feature Importance (Neural Network Weights)', fontsize=16)
plt.xlabel('Average Weight Magnitude')
plt.ylabel('Features')
plt.show()

```

```python
predictions = model.predict(X_test)
predicted_classes = np.argmax(predictions, axis=1) + 1  
```
